```python
# Import dataloaders
import dataset.cifar10 as cifar10
import dataset.svhn as svhn
import dataset.cifar10_c as cifar10_c
```

```python
# Import network architectures
from module.resnet import resnet50, resnet110
from module.wide_resnet import wide_resnet_cifar
from module.densenet import densenet121
```

```python
# Import metrics to compute
from Metrics.ood_test_utils import get_roc_auc
```

```python
# Import plot related libraries
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
# Dataset params
dataset_num_classes = {
    'cifar10': 10,
    'svhn': 10,
    'cifar10_c': 10
}

dataset_loader = {
    'cifar10': cifar10,
    'cifar10_c': cifar10_c,
    'svhn': svhn
}
```

```python
# Mapping model name to model function
models = {
    'resnet50': resnet50,
    'resnet110': resnet110,
    'wide_resnet': wide_resnet_cifar,
    'densenet121': densenet121,
}
```

```python
# Checking if GPU is available
cuda = False
if (torch.cuda.is_available()):
    cuda = True

# Setting additional parameters
torch.manual_seed(1)
device = torch.device("cuda" if cuda else "cpu")
```

```python
class args:
    data_aug = True
    gpu = device == "cuda"
    train_batch_size = 128
    test_batch_size = 128
```

```python
dataset = 'cifar10'
ood_dataset = 'svhn'

num_classes = dataset_num_classes[dataset]
train_loader, val_loader = dataset_loader[dataset].get_train_valid_loader(
    batch_size=args.train_batch_size,
    augment=args.data_aug,
    random_seed=1,
    pin_memory=args.gpu
)

test_loader = dataset_loader[dataset].get_test_loader(
    batch_size=args.test_batch_size,
    pin_memory=args.gpu
)


ood_train_loader, ood_val_loader = dataset_loader[ood_dataset].get_train_valid_loader(
    batch_size=args.train_batch_size,
    augment=args.data_aug,
    random_seed=1,
    pin_memory=args.gpu
)

ood_test_loader = dataset_loader[ood_dataset].get_test_loader(
    batch_size=args.test_batch_size,
    pin_memory=args.gpu
)
```

```python
# hfai workspace download checkpoints/Search_Retrain11050540_CEFormat_cifar10_resnet50_num_states=16-2/retrain_latest.pt
# hfai workspace download checkpoints/Search_Retrain11050540_CEFormat_cifar10_resnet110_num_states=18-2/retrain_latest.pt
# hfai workspace download checkpoints/Search_Retrain11041114_noCEFormat_cifar10_wide_resnet_num_states=20-0/retrain_latest.pt
# hfai workspace download checkpoints/Search_Retrain11041114_noCEFormat_cifar10_densenet121_num_states=15-4/retrain_latest.pt


# hfai workspace download checkpoints/Search_Retrain11041114_noCEFormat_cifar100_resnet50_num_states=19-0/retrain_latest.pt
# hfai workspace download checkpoints/Search_Retrain11050540_noCEFormat_cifar100_resnet110_num_states=16-4/retrain_latest.pt
# hfai workspace download checkpoints/Search_Retrain11041114_noCEFormat_cifar100_wide_resnet_num_states=20-0/retrain_latest.pt
# hfai workspace download checkpoints/Search_Retrain11050540_noCEFormat_cifar100_densenet121_num_states=15-2/retrain_latest.pt

```

```python

save_dir = {
    'resnet50': "checkpoints/Search_Retrain11050540_CEFormat_cifar10_resnet50_num_states=16-2",
    'resnet110': "checkpoints/Search_Retrain11050540_CEFormat_cifar10_resnet110_num_states=18-2",
    'wide_resnet': "checkpoints/Search_Retrain11041114_noCEFormat_cifar10_wide_resnet_num_states=20-0",
    'densenet121': "checkpoints/Search_Retrain11041114_noCEFormat_cifar10_densenet121_num_states=15-4",
}
T_opt = {
    'resnet50': 3.3,
    'resnet110': 3.1,
    'wide_resnet': 2.2,
    'densenet121': 2.4,
}
```

```python
# Taking input for the model
model_name = "wide_resnet"
model = models[model_name]
```

```python


net = model(num_classes=num_classes, temp=1.0)
net.cuda()
cudnn.benchmark = True
ckpt = torch.load(f"{save_dir[model_name]}/retrain_latest.pt", map_location='cpu')
net.load_state_dict(ckpt['model'])

(fpr_entropy, tpr_entropy, thresholds_entropy), (fpr_confidence, tpr_confidence, thresholds_confidence), auc_entropy, auc_confidence = get_roc_auc(net, test_loader, ood_test_loader, device)

clrs = ['#1f77b4','#ff7f0e', '#2ca02c','#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22','#17becf']

plt.figure()
plt.rcParams["figure.figsize"] = (10, 8)
sb.set_style('whitegrid')
plt.plot(fpr_entropy, tpr_entropy, color=clrs[0], linewidth=5, label='ROC')

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.xlabel('FPR', fontsize=30)
plt.ylabel('TPR', fontsize=30)
plt.legend(fontsize=28)

print ('AUROC entropy: ' + str(auc_entropy))
```

```python
net = model(num_classes=num_classes, temp=T_opt[model_name])
net.cuda()
cudnn.benchmark = True
ckpt = torch.load(f"{save_dir[model_name]}/retrain_latest.pt", map_location='cpu')
net.load_state_dict(ckpt['model'])

(fpr_entropy, tpr_entropy, thresholds_entropy), (fpr_confidence, tpr_confidence, thresholds_confidence), auc_entropy, auc_confidence = get_roc_auc(net, test_loader, ood_test_loader, device)

clrs = ['#1f77b4','#ff7f0e', '#2ca02c','#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22','#17becf']

plt.figure()
plt.rcParams["figure.figsize"] = (10, 8)
sb.set_style('whitegrid')
plt.plot(fpr_entropy, tpr_entropy, color=clrs[0], linewidth=5, label='ROC')

plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.xlabel('FPR', fontsize=30)
plt.ylabel('TPR', fontsize=30)
plt.legend(fontsize=28)

print ('AUROC entropy: ' + str(auc_entropy))
```
